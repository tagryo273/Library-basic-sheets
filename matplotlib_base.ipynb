```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
```

#基本的な2つの書き方
  1つ目は簡易的な方法で、少ないコードで済むが自由度が高くない
  2つ目は、もう少し厳格な方法で、カスタマイズ可能だが、その分きちんとコードを書かなくてはいけない

簡易的な方法

```
x=np.linspace(0,1,10)#0から1を10個に分割
y = 2*x+0.1*np.random.randn(10)
plt.plot(x,y)
plt.show()
```

厳格な方法

```
x = np.linspace(0,1,10)
y = 2*x+0.1*np.random.randn(10)
```

```
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y)
plt.show()
```

figure()とadd_subsplot()メソッドを使って生成する違い
>figとaxはそれぞれ、インターネットブウザのウィンドとタブのような関係。つまりfigグラフは配置する領域の事で、axはグラフを描画する領域を示している。
>add_subplot(111)は、1行1列(つまり1個のグラフ)のうち1番目に、グラフを描画
する領域を追加している。

複数のグラフを描画する

```
x_1 = np.linspace(0,1,10)
y_1 = x_1+0.1*np.random.randn(10)

x_2 = np.linspace(0,1,100)
y_2 =2*x_2+0.1*np.random.randn(100)

x_3 = np.linspace(0,1,1000)
y_3 = 3*x_3+0.1*np.random.randn(1000)
```

```
fig = plt.figure()

ax_1 = fig.add_subplot(221)#2行2列1番目
ax_2 = fig.add_subplot(222)#2行2列2番目
ax_3 = fig.add_subplot(223)#2行2列3番目

ax_1.plot(x_1,y_1)
ax_2.plot(x_2,y_2)
ax_3.plot(x_3,y_3)

plt.show()
```

#グラフの大きさを操作する
グラフの大きさを変えるには、fig生成時にfigsizeを指定する


```
fig = plt.figure(figsize=(20,12))#横20縦12

ax_1 = fig.add_subplot(221)
ax_2 = fig.add_subplot(222)
ax_3 = fig.add_subplot(223)

ax_1.plot(x_1,y_1)
ax_2.plot(x_2,y_2)
ax_3.plot(x_3,y_3)

plt.show()
```

簡易版での書き方

```
x_1 = np.linspace(0,1,10)
y_1 = x_1+0.1*np.random.randn(10)

plt.figure(figsize=(12,8))
plt.plot(x_1,y_1)
plt.show()
```

#figureとaxの生成を同時に行う。
figとaxを同時に生成するには、subplots()メソッドを使う。

```
fig,axes = plt.subplots(2,2,figsize=(20,12)) 
```

```
axes
```

```
axes.shape
```

```
fig,axes = plt.subplots(2,2,figsize=(20,12)) 
axes[0][0].plot(x_1,y_1)
axes[0][1].plot(x_2,y_2)
axes[1][0].plot(x_3,y_3)
plt.show()
```

# axesを一次元配列に変換
for文などでループさせる際に行、列どちらも指定する2次元配列は不便
である。
>ravel()メソッドでaxesを1次元に変換すれば、通し番号だいえでグラフを指定することができる。


```
fig,axes = plt.subplots(2,2,figsize=(20,12))

axes.ravel()[0].plot(x_1,y_1)
axes.ravel()[1].plot(x_2,y_2)
axes.ravel()[2].plot(x_3,y_3)

plt.show()
```

#散布図
散布図を描画には、scatterメソッドを使う。

```
df = pd.DataFrame(data=np.random.randn(500,2),columns=['C1','C2'])
df.head()
```

```
plt.figure(figsize=(12,8))
plt.scatter(df['C1'],df['C2'])#plt.scatter(x,y)
plt.show()
```

点の大きさ・色・スタイルを変更する

```
plt.figure(figsize=(12,8))
plt.scatter(df['C1'],df['C2'],s=200,c='#f7d945',marker='^',alpha=0.5)
#s大きさ　c色　marker形 alpha透明度
plt.title('test',fontsize=20)
plt.xlabel('$x_0$',fontsize=15)
plt.ylabel('$x_1$',fontsize=15)
plt.show()
```

**点の大きさや色をデータによって変える**

```
df['data_size'] = np.random.randint(1,100,500)#１から100を500個
df.head()
```

```
plt.figure(figsize=(12,8))
plt.scatter(df['C1'],df['C2'],s=df['data_size']*10,c=df['data_size'],cmap='viridis',alpha=0.8)
#s大きさ　c色　marker形 alpha透明度 data_sizeによっていろや大きさを変えた
plt.colorbar()
plt.title('test',fontsize=20)
plt.xlabel('$x_0$',fontsize=15)
plt.ylabel('$x_1$',fontsize=15)
plt.show()
```

#棒グラフ
棒グラフを描画するには、bar()メソッドを使う。

```
df = pd.DataFrame({'C1':['A','A','A','B','B','C'],
                   'C2':[20,50,60,80,100,30],
                   'C3':[40,200,100,500,40,200]})
df
```

```
df['C1'].value_counts().index
```

```
df.groupby('C1').sum()#C1が同じものを合計した
```

```
plt.figure(figsize=(12,8))
plt.bar(df['C1'].value_counts().index, df.groupby('C1').sum()['C2'])#plt.bar(x,y)
                #A,B,Cを取り出した
                #xが棒を立てる位置 yが棒の高さ
plt.show()
```

複数の棒グラフを描画

```
df=pd.DataFrame(data=100*np.random.rand(5,5),index=['A','B','C','D','E'],
                columns=['C1','C2','C3','C4','C5'])
df.head()
```

```
plt.figure(figsize=(12,8))

x = np.arange(len(df))

bar_width=0.4

plt.bar(x , df['C1'] , color='red' , width=bar_width, label='C1')
plt.bar(x+bar_width , df['C2'] , color='blue' , width=bar_width, label='C2')

plt.xticks(x+bar_width / 2,df.index,fontsize=20)
plt.legend()
plt.show()
```

```
plt.figure(figsize=(12,8))

x = np.arange(len(df), dtype='float64')

bar_width=0.1

for col in df.columns:
  plt.bar(x,df[col],width=bar_width,label=col)
  x +=bar_width

plt.xticks((np.arange(len(df))+x+bar_width) / 2,df.index,fontsize=20)
plt.legend()
plt.show()
```

#hヒストグラム
ヒストグラムの描画には、histメソッドを使う

```
df = pd.DataFrame(data=np.random.randn(500), columns=['C1'])
df.head()
```

```
plt.figure(figsize=(12,8))
plt.hist(df['C1'])
plt.show()
```

見た目を整える
>ビンの数を指定するbins,棒の幅を指定しているrwidthなどを調節するとより見やすいヒストグラムになる

```
plt.figure(figsize=(12,8))
plt.hist(df['C1'], bins=20, rwidth=0.8,color='#63738c')
plt.show()
```

```
from scipy.stats import norm

def histogram(x,bins=20):
  plt.figure(figsize=(12,8))
  plt.hist(df['C1'], bins=20, rwidth=0.8,color='#63738c',density=True)

  mean = x.mean()
  std = x.std()

  X = np.linspace(x.min(),x.max(),bins)
  y = norm.pdf(X,mean,std)

  plt.plot(X , y , linewidth=3)
  plt.show()

histogram(df['C1'])
```

#円グラフ
円グラフの描画には,pieメソッドを使う。

```
df = pd.DataFrame(data=np.random.randint(0,5,500),columns=['C1'])
df.head()
```

```
df['C1'].value_counts()
```

```
plt.figure(figsize=(10,10))
plt.pie(df['C1'].value_counts(), labels=df['C1'].value_counts().index)
plt.show()
```

見た目を整える

```
plt.figure(figsize=(10,10))

testprops = {'fontsize':20}
plt.pie(df['C1'].value_counts(),
        labels=df['C1'].value_counts().index,
        autopct='%.1f%%',#割合を表示
        startangle=90,#回転
        counterclock=False,#時計回りに
        textprops=testprops)
plt.show()
```
